```python
average=[i-1 for i in average]
inbreast=[i-1 for i in inbreast]
bcdr = [i-1 for i in bcdr]
stge = [i-1 for i in stge]
jarv = [i-1 for i in jarv]
cmmd = [i-1 for i in cmmd]
```
